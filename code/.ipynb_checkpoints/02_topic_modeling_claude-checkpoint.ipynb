Covers: topic identification, summary table, relationship tree, and STOPWORDS revision.

## 2. Load Data

````python
df = pd.read_csv("../data/01_filtered_cases.csv")
print("Dataset shape:", df.shape)
print("Columns:", df.columns.tolist())
df.head(3)
````

## 3. Parse LDA Visualization HTML

````python
with open("../output/lda_vis_chn.html") as f:
    html = f.read()

match = re.search(r'var ldavis_\w+_data = ({.*?});', html, re.DOTALL)
lda_data = json.loads(match.group(1))
tinfo    = lda_data["tinfo"]
mds      = lda_data["mdsDat"]

print("Topic order (display -> gensim 1-indexed):", lda_data["topic.order"])
print("\nDisplay topic frequencies:")
for t, f in zip(mds["topics"], mds["Freq"]):
    print(f"  Display Topic {t}: {f:.2f}%")
````

## 4. Top Terms per Topic (by Loglift & Frequency)

````python
topic_terms = {}
for tn in range(1, 6):
    cat = f"Topic{tn}"
    rows = [(tinfo["Term"][i], tinfo["Freq"][i], tinfo["loglift"][i])
            for i, c in enumerate(tinfo["Category"]) if c == cat]
    by_ll   = sorted(rows, key=lambda x: -x[2])[:12]
    by_freq = sorted(rows, key=lambda x: -x[1])[:12]
    topic_terms[tn] = {"loglift": by_ll, "freq": by_freq}
    print(f"=== Display Topic {tn} | top by loglift ===")
    print("  ", [t for t, f, ll in by_ll])
    print(f"  top by freq:")
    print("  ", [t for t, f, fr in by_freq])
    print()
````

## 7. Inter-Topic Relationships

| Relationship | Topics | Shared Signal |
|---|---|---|
| Both center on **athlete employment & compensation** | 1 ↔ 3 | 退役费, 补发, 体育局 |
| Both involve **financial quantification** | 3 ↔ 5 | 计付, 欠款, 损失 |
| Both have **commercial/contract** framing | 2 ↔ 5 | 欠付, 履行合同, 转款 |
| Both use **WeChat / digital evidence** | 4 ↔ 5 | 微信, 截图, 聊天记录 |
| Shared **sports-governance** vocabulary | 1 ↔ 3 | 体育局, 运动队, 国家体育总局 |

**Structural clusters:**
- **Sports-specific cluster** → T1, T3, T4: athlete as direct party; sport-specific facts dominate
- **Commercial / financial periphery** → T2, T5: parties may be clubs/organisations; less sport-specific
- T1 acts as a **hub**: overlaps with T2 (admin/contract), T3 (compensation), and sports-governance vocabulary in T4


## 8. Topic Relationship Tree Graph

````python
fig, ax = plt.subplots(figsize=(13, 8))
ax.set_xlim(-0.5, 5.5)
ax.set_ylim(-0.5, 4.5)
ax.axis("off")
ax.set_facecolor("#f8f9fa")
fig.patch.set_facecolor("#f8f9fa")

COLORS = {
    "root":          "#2C3E50",
    "branch_sports": "#1A6A9E",
    "branch_comm":   "#7D3C98",
    1: "#1A9E6A",
    3: "#E67E22",
    4: "#C0392B",
    2: "#8E44AD",
    5: "#2980B9",
}
SIZES = {1: 2200, 2: 1300, 3: 900, 4: 800, 5: 400}

NODE_POS = {
    "root":       (2.5, 4.0),
    "sports":     (1.2, 2.8),
    "commercial": (3.8, 2.8),
    1: (0.3, 1.4),
    3: (1.2, 1.4),
    4: (2.1, 1.4),
    2: (3.2, 1.4),
    5: (4.4, 1.4),
}

LABELS = {
    "root":       "Athlete\nCourt Cases",
    "sports":     "Sports-Specific\nDisputes",
    "commercial": "Commercial /\nFinancial",
    1: "T1 (41.5%)\n运动员退役\n就业补偿",
    3: "T3 (15.2%)\n运动伤害\n医疗赔偿",
    4: "T4 (13.5%)\n网络侵权\n肖像权",
    2: "T2 (23.4%)\n商业合同\n违约",
    5: "T5 (6.5%)\n借贷债务\n追偿",
}

tree_edges = [
    ("root", "sports"),
    ("root", "commercial"),
    ("sports", 1), ("sports", 3), ("sports", 4),
    ("commercial", 2), ("commercial", 5),
]
cross_edges = [(1, 3), (3, 5), (2, 5), (4, 5)]

for a, b in tree_edges:
    x0, y0 = NODE_POS[a]; x1, y1 = NODE_POS[b]
    ax.plot([x0, x1], [y0, y1], color="#666", lw=2.0, alpha=0.8, zorder=1)

for a, b in cross_edges:
    x0, y0 = NODE_POS[a]; x1, y1 = NODE_POS[b]
    ax.plot([x0, x1], [y0, y1], color="#aaa", lw=1.2,
            linestyle="--", alpha=0.55, zorder=1)

for key, (x, y) in NODE_POS.items():
    if key == "root":
        s, c, tc = 2800, COLORS["root"], "white"
    elif key in ("sports", "commercial"):
        s = 1800
        c = COLORS["branch_sports"] if key == "sports" else COLORS["branch_comm"]
        tc = "white"
    else:
        s, c, tc = SIZES[key], COLORS[key], "white"
    ax.scatter(x, y, s=s, c=c, zorder=3, edgecolors="white", linewidths=2)
    ax.text(x, y, LABELS[key], ha="center", va="center", fontsize=7.5,
            color=tc, fontweight="bold", zorder=4, linespacing=1.4)

legend_handles = [
    mpatches.Patch(color=COLORS[1], label="T1 – Retirement / Employment"),
    mpatches.Patch(color=COLORS[3], label="T3 – Sports Injury / Medical"),
    mpatches.Patch(color=COLORS[4], label="T4 – Image Rights / Online"),
    mpatches.Patch(color=COLORS[2], label="T2 – Commercial Contract"),
    mpatches.Patch(color=COLORS[5], label="T5 – Debt / Lending"),
    mpatches.Patch(color="#aaa", label="- -  Topic overlap (shared terms)"),
]
ax.legend(handles=legend_handles, loc="lower left", fontsize=8,
          framealpha=0.85, edgecolor="#ccc")
ax.set_title(
    "LDA Topic Relationship Tree\n"
    "(node size proportional to corpus share; dashed = shared vocabulary)",
    fontsize=11, fontweight="bold", pad=12)
plt.tight_layout()
plt.savefig("../output/topic_tree.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved -> ../output/topic_tree.png")
````

## 9. Proposed STOPWORDS Modifications

````python
STOPWORDS_ORIGINAL = set("""
一审 二审 再审 原告 被告 申请人 被申请人 上诉人 被上诉人 驳回上诉 诉至 抗诉 审判长 审判员
书记员 院长 副院长 庭长 法官 助理 合议庭 人民法院 高级人民法院 中级人民法院 基层人民法院
判决 裁定 裁决 判决书 裁定书 法律 如下 认为 依照 根据 查明 本院 经审查 经审理 经查 裁决书 检察院 检察员
中华人民共和国 人民币 甲方 乙方 丙方 甲乙双方 本院认为 无证据 诉讼法 律师费 民法典 公告费 不服 传唤 案外人 给付
举办方 本院认为 原审 贵院 实缴 该证据 受理费 住所地 辩称 上诉状 判令 载明 诉讼请求 简易程序 采信 起诉 公证费
审批表 诉称 综上 副本 举证 自认 诉请 人民陪审员 开庭 受案 上述事实 本合同 应予 处理结果 不予 提出 证据
答辩 维持原判 驳回 纠纷 驳回 委托 审查 被执行人 撤诉 答辩 终号 因案 维持原判 发回重审 终审 案号 诉讼费用 在案
""".split())

STOPWORDS_ORIGINAL.update(
    "委托, 诉讼, 代理人, 律师, 事务所, 法定代表, 事务所律师, 上诉, 立案, 依法, 组成, 进行, 审理,"
    " 本案, 现已, 终结, 请求, 依法, 一审判决, 第一, 判项, 改判, 诉讼费, 事实, 理由, 承担,"
    " 理由, 法院, 最高人民法院, 关于, 若干, 问题, 指导, 意见, 第三条, 第二项, 原名, 称为, 不可, 不遗余力, 后于"
    .split(", "))

print(f"Original STOPWORDS size: {len(STOPWORDS_ORIGINAL)}")
````

````python
# Terms to ADD — appeared in LDA but are procedural/structural boilerplate
TO_ADD = {
    "民事判决":  "document-type label — top-frequency in T3/T5; present across all topics as header text",
    "民事裁定":  "document-type label — top word in Gensim topic 3 (Display T1); pure case header",
    "民事":      "generic civil prefix — top frequency in multiple topics; zero discriminating power",
    "出生":      "birth-date field in case header (appeared in Display T1 frequency top)",
    "汉族":      "ethnicity field in case header (appeared in Display T1 frequency top)",
    "送达":      "service of process — procedural; 0.020 weight in Display T4 inflates topic",
    "公告送达":  "public service of process — procedural boilerplate in Display T4",
    "案件":      "generic case — no topic discrimination; dominated Gensim topic 3",
    "申请":      "generic application — procedural; mixed into Display T1 high-freq list",
    "规定":      "generic regulations — too broad; dilutes T1 coherence",
    "京民":      "Beijing civil case-number prefix — artifact from case ID string parsing",
}

print("Terms proposed to ADD to STOPWORDS:")
print(f"  Count: {len(TO_ADD)}")
print()
for term, reason in TO_ADD.items():
    print(f"  + {term:<12} {reason}")
````

````python
STOPWORDS_MODIFIED = STOPWORDS_ORIGINAL | set(TO_ADD.keys())

print(f"Original size : {len(STOPWORDS_ORIGINAL)}")
print(f"Modified size : {len(STOPWORDS_MODIFIED)}")
print(f"Net additions : {len(STOPWORDS_MODIFIED) - len(STOPWORDS_ORIGINAL)}")

print("\n=== CHANGE LOG ===")
print(f"{'Action':<8} {'Term':<14} Rationale")
print("-" * 80)
print("  (no removals — all original terms retain validity)")
print()
for term, reason in TO_ADD.items():
    print(f"  ADD    {term:<14} {reason[:62]}")
````

## 11. Sanity Check — New Terms in Raw Corpus

````python
import re as _re

counts = {}
for term in TO_ADD:
    cnt   = df["full_text"].str.count(_re.escape(term)).sum()
    n_doc = df["full_text"].str.contains(_re.escape(term)).sum()
    counts[term] = {"occurrences": int(cnt), "n_docs": int(n_doc),
                    "pct_docs": round(int(n_doc) / len(df) * 100, 1)}

freq_df = (pd.DataFrame(counts).T
           .rename(columns={"occurrences": "Raw Occurrences",
                            "n_docs": "Docs Containing",
                            "pct_docs": "% of Corpus"})
           .sort_values("Raw Occurrences", ascending=False))
print(freq_df.to_string())
````

````python
fig, ax = plt.subplots(figsize=(9, 4))
bars = ax.barh(freq_df.index, freq_df["Raw Occurrences"].astype(int),
               color="#C0392B", alpha=0.8)
ax.bar_label(bars, padding=3, fontsize=8)
ax.set_xlabel("Raw term occurrences across all case texts")
ax.set_title(
    "Proposed new STOPWORDS — raw corpus frequency\n"
    "(high counts confirm these are boilerplate, not substantive topic signals)")
ax.invert_yaxis()
plt.tight_layout()
plt.savefig("../output/new_stopwords_freq.png", dpi=130, bbox_inches="tight")
plt.show()
````
