# 04 — Keyword Density (Full-Text Corpus)

Visualises how the **top 20 most frequent keywords across the entire case corpus** distribute across the five LDA topics identified in `03_topic_summary.ipynb`.

**Key difference from `04_keyword_density_modified.ipynb`:**
- `_modified` selects terms by LDA loglift (most *discriminative* per topic)
- This notebook selects terms by raw **corpus-wide document frequency** from the DTM (`02_dtm.ipynb`), then maps each term to its topic affinity via pyLDAvis `token.table`

Terms marked with † are boilerplate survivors (民事, 民事判决, 民事裁定) that were added to the modified STOPWORDS after the initial LDA review.

## 1. Imports & Font

```python
import json, re, ast
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.colors as mcolors
import matplotlib.ticker
import matplotlib.font_manager as fm
from matplotlib.gridspec import GridSpec
from sklearn.feature_extraction.text import CountVectorizer

_cjk = ['PingFang SC', 'Heiti SC', 'STHeiti', 'SimHei', 'Noto Sans CJK SC']
_avail = {f.name for f in fm.fontManager.ttflist}
_font  = next((f for f in _cjk if f in _avail), None)
if _font:
    matplotlib.rcParams['font.family'] = _font
    print(f'Using CJK font: {_font}')
else:
    print('No CJK font — Chinese labels may render as boxes')
matplotlib.rcParams['axes.unicode_minus'] = False

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

## 2. Build DTM & Extract Top 20 Corpus Terms

Replicates the `02_dtm.ipynb` pipeline: `CountVectorizer` on the string-serialized keyword lists in `02_tokenized.csv`, then sums columns to rank terms by how many documents contain each keyword.

```python
df = pd.read_csv("../data/02_tokenized.csv")
print(f"Corpus: {df.shape[0]} cases")

corpus = df.keywords.to_list()

vectorizer = CountVectorizer(lowercase=True)
dtm_sparse = vectorizer.fit_transform(corpus)
dtm = pd.DataFrame(dtm_sparse.todense(),
                   columns=vectorizer.get_feature_names_out())

top_terms_series = dtm.sum(axis=0).sort_values(ascending=False)
top20_terms = top_terms_series.head(20).index.tolist()
top20_counts = top_terms_series.head(20).astype(int)

print("\nTop 20 corpus-wide keywords (document frequency):")
for i, (t, c) in enumerate(top20_counts.items(), 1):
    boiler = " †" if t in {"民事", "民事判决", "民事裁定"} else ""
    print(f"  {i:2d}. {t:<14}{c:>4} docs{boiler}")
print("\n† = boilerplate term added to modified STOPWORDS")
```

## 3. Load LDA Topic Weights from `lda_vis_chn.html`

For each of the 20 corpus-top terms, retrieve p(topic | term) from pyLDAvis's `token.table`. This shows how the LDA model distributes each term's probability mass across the five topics.

```python
with open("../output/lda_vis_chn.html") as f:
    html = f.read()

match    = re.search(r'var ldavis_\w+_data = ({.*?});', html, re.DOTALL)
lda_data = json.loads(match.group(1))
tt       = lda_data["token.table"]

df_tt = pd.DataFrame({"term": tt["Term"], "topic": tt["Topic"], "prob": tt["Freq"]})
pivot = df_tt.pivot_table(index="term", columns="topic",
                           values="prob", aggfunc="sum", fill_value=0)
pivot.columns = [f"T{c}" for c in pivot.columns]

# Build the 20 × 5 weight matrix, preserving frequency rank order
matrix = pivot.reindex(top20_terms, fill_value=0)
topics = ["T1", "T2", "T3", "T4", "T5"]
Z = matrix[topics].values

print("Weight matrix  p(topic | term):")
print(matrix[topics].round(3).to_string())
```

## 4. Topic Colours, Labels & Row Grouping

```python
COLORS = {
    "T1": "#8E44AD",  # purple  — Commercial Contract & Debt
    "T2": "#C0392B",  # red     — Online Infringement / Image Rights
    "T3": "#1A9E6A",  # green   — Sports Personnel & Wages
    "T4": "#E67E22",  # orange  — Retirement & Administrative Litigation
    "T5": "#2980B9",  # blue    — Sports Injury & Medical Compensation
    "none": "#AAAAAA",
}

TOPIC_LABELS = {
    "T1": "T1\n商业合同违约\n与债务追偿",
    "T2": "T2\n网络侵权\n肖像权纠纷",
    "T3": "T3\n运动队人事\n薪资待遇",
    "T4": "T4\n退役安置\n行政诉讼",
    "T5": "T5\n运动伤害\n医疗赔偿",
}
TOPIC_PCT = {"T1": 23.9, "T2": 23.9, "T3": 22.8, "T4": 19.7, "T5": 9.7}

BOILERPLATE = {"民事", "民事判决", "民事裁定"}

# Dominant topic for each term = topic with highest p(topic|term)
def dominant(row):
    vals = {t: row[t] for t in topics}
    best = max(vals, key=vals.get)
    return best if vals[best] > 0 else "none"

ROW_DOM = {term: dominant(matrix.loc[term]) for term in top20_terms}

print("Dominant topic per corpus-top-20 term:")
for term in top20_terms:
    flag = " †" if term in BOILERPLATE else ""
    print(f"  {term:<14} dominant={ROW_DOM[term]}  "
          f"weights={matrix.loc[term, topics].round(2).to_dict()}{flag}")
```

## 5. Keyword Density Heatmap

Rows = top 20 corpus terms in frequency rank (most frequent at top). Cell colour intensity = p(topic | keyword). Left-side tick colour = dominant topic. Right-side bar = document count from DTM (n out of 79 cases). † marks boilerplate terms that were filtered out in the modified STOPWORDS.

```python
n_terms  = len(top20_terms)
n_topics = 5

fig = plt.figure(figsize=(13, 10))
fig.patch.set_facecolor("#f8f9fa")
gs  = GridSpec(1, 2, width_ratios=[13, 1], wspace=0.04, figure=fig)
ax  = fig.add_subplot(gs[0])
axb = fig.add_subplot(gs[1])
ax.set_facecolor("#f8f9fa"); axb.set_facecolor("#f8f9fa")

# ── cells ──────────────────────────────────────────────────────────────────
for col_idx, tc in enumerate(topics):
    base = mcolors.to_rgb(COLORS[tc])
    for row_idx, term in enumerate(top20_terms):
        val = Z[row_idx, col_idx]
        cell_color = tuple(1 - val * (1 - c) for c in base)
        ax.add_patch(plt.Rectangle(
            [col_idx - 0.5, row_idx - 0.5], 1, 1,
            facecolor=cell_color, edgecolor="white", linewidth=1.2))
        if val > 0.07:
            txt_c = "white" if val > 0.55 else "#2C3E50"
            ax.text(col_idx, row_idx, f"{val:.2f}",
                    ha="center", va="center", fontsize=8,
                    color=txt_c, fontweight="bold")

# ── separators every 5 rows (visual rhythm for rank groups) ────────────────
for sep in [4.5, 9.5, 14.5]:
    ax.axhline(sep, color="#999", linewidth=1.0, linestyle=":", alpha=0.7)

# ── y-axis: term labels coloured by dominant topic ─────────────────────────
ax.set_ylim(-0.5, n_terms - 0.5)
ax.set_yticks(range(n_terms))
ylabels = [f"{t} †" if t in BOILERPLATE else t for t in top20_terms]
ax.set_yticklabels(ylabels, fontsize=10.5)
for tick, term in zip(ax.get_yticklabels(), top20_terms):
    dom = ROW_DOM[term]
    tick.set_color(COLORS[dom])
    tick.set_fontweight("bold" if term not in BOILERPLATE else "normal")
ax.invert_yaxis()

# ── x-axis: topic column headers ────────────────────────────────────────────
ax.set_xlim(-0.5, n_topics - 0.5)
ax.set_xticks(range(n_topics))
ax.set_xticklabels([TOPIC_LABELS[t] for t in topics], fontsize=8.5, linespacing=1.4)
for tick, tc in zip(ax.get_xticklabels(), topics):
    tick.set_color(COLORS[tc]); tick.set_fontweight("bold")
ax.xaxis.set_tick_params(length=0)
ax.xaxis.tick_top()
ax.tick_params(axis="y", length=0)
for spine in ax.spines.values(): spine.set_visible(False)

# ── frequency rank labels on the far left ──────────────────────────────────
for row_idx, term in enumerate(top20_terms):
    ax.text(-0.7, row_idx, f"#{row_idx+1}",
            ha="center", va="center", fontsize=7.5, color="#999")

# ── right-side bar: raw document count ─────────────────────────────────────
counts = [int(top20_counts[t]) for t in top20_terms]
bar_colors = [COLORS[ROW_DOM[t]] for t in top20_terms]
axb.barh(range(n_terms), counts, color=bar_colors, alpha=0.7,
         edgecolor="white", linewidth=0.8, height=0.65)
axb.set_xlim(0, max(counts) * 1.15)
axb.set_ylim(-0.5, n_terms - 0.5)
axb.invert_yaxis()
axb.set_xticks([0, 15, 30])
axb.set_xticklabels(["0", "15", "30"], fontsize=7, color="#666")
axb.set_yticks([])
axb.set_xlabel("Docs\n(n=79)", fontsize=7, color="#666", labelpad=2)
axb.tick_params(axis="x", length=2, colors="#999")
for spine in ["top", "right", "left"]: axb.spines[spine].set_visible(False)
axb.spines["bottom"].set_color("#ccc")
for sep in [4.5, 9.5, 14.5]:
    axb.axhline(sep, color="#999", linewidth=1.0, linestyle=":", alpha=0.5)

# ── legend ──────────────────────────────────────────────────────────────────
handles = []
for tc in topics:
    lbl = TOPIC_LABELS[tc].replace("\n", " ") + f" ({TOPIC_PCT[tc]:.1f}%)"
    handles.append(mpatches.Patch(color=COLORS[tc], label=lbl))
handles.append(mpatches.Patch(color="#AAAAAA", label="No dominant topic (T1 = 0 for all)"))
ax.legend(handles=handles, loc="lower center", ncol=3, fontsize=7.5,
          framealpha=0.9, edgecolor="#ccc", bbox_to_anchor=(0.5, -0.09))

fig.suptitle(
    "Keyword Density across LDA Topics — Top 20 Corpus-Wide Terms\n"
    "Cell intensity = p(topic | keyword)  ·  rank # = corpus frequency rank  ·  † = boilerplate (modified STOPWORDS)",
    fontsize=10.5, fontweight="bold", y=1.02
)
plt.savefig("../output/keyword_density_fulltext_heatmap.png",
            dpi=150, bbox_inches="tight")
plt.show()
print("Saved → ../output/keyword_density_fulltext_heatmap.png")
```

## 6. Stacked Topic Distribution Bars

Each horizontal bar is one of the 20 terms, normalised so segments sum to 100%. Terms where T1 (purple) is absent confirm that commercial/debt vocabulary does not overlap with the highest-frequency corpus terms.

Rows remain in **corpus frequency rank order** (most frequent at top).

```python
fig2, ax2 = plt.subplots(figsize=(11, 9))
fig2.patch.set_facecolor("#f8f9fa"); ax2.set_facecolor("#f8f9fa")

row_sums = Z.sum(axis=1, keepdims=True)
row_sums[row_sums == 0] = 1
Z_norm = Z / row_sums

y_pos = np.arange(n_terms)
lefts = np.zeros(n_terms)

for col_idx, tc in enumerate(topics):
    widths = Z_norm[:, col_idx]
    ax2.barh(y_pos, widths, left=lefts,
             color=COLORS[tc], edgecolor="white", linewidth=0.8, height=0.65,
             label=TOPIC_LABELS[tc].replace("\n", " "))
    for i, (w, l) in enumerate(zip(widths, lefts)):
        if w > 0.10:
            ax2.text(l + w / 2, i, f"{w:.0%}",
                     ha="center", va="center", fontsize=7.5,
                     color="white", fontweight="bold")
    lefts = lefts + widths

# rank-group separators
for sep in [4.5, 9.5, 14.5]:
    ax2.axhline(sep, color="#999", linewidth=1.0, linestyle=":", alpha=0.6)

# y-axis
ax2.set_yticks(y_pos)
ax2.set_yticklabels(ylabels, fontsize=10.5)
for tick, term in zip(ax2.get_yticklabels(), top20_terms):
    dom = ROW_DOM[term]
    tick.set_color(COLORS[dom])
    tick.set_fontweight("bold" if term not in BOILERPLATE else "normal")
ax2.invert_yaxis()

# rank labels
for row_idx in range(n_terms):
    ax2.text(-0.05, row_idx, f"#{row_idx+1}",
             ha="right", va="center", fontsize=7.5, color="#999",
             transform=ax2.get_yaxis_transform())

ax2.set_xlim(0, 1.0)
ax2.set_xlabel("Proportion of keyword probability assigned to each LDA topic", fontsize=9)
ax2.xaxis.set_major_formatter(matplotlib.ticker.PercentFormatter(xmax=1))
ax2.tick_params(axis="x", labelsize=8); ax2.tick_params(axis="y", length=0)
for spine in ["top", "right"]: ax2.spines[spine].set_visible(False)
ax2.spines["left"].set_color("#ccc"); ax2.spines["bottom"].set_color("#ccc")

handles2 = [mpatches.Patch(color=COLORS[tc],
            label=TOPIC_LABELS[tc].replace("\n"," ") + f" ({TOPIC_PCT[tc]:.1f}%)")
            for tc in topics]
ax2.legend(handles=handles2, loc="lower right", fontsize=8,
           framealpha=0.9, edgecolor="#ccc", ncol=2)

ax2.set_title(
    "Top 20 Corpus Terms — Topic Distribution (Stacked)\n"
    "Rows in corpus frequency rank · † = boilerplate (modified STOPWORDS)",
    fontsize=11, fontweight="bold", pad=10
)
plt.tight_layout()
plt.savefig("../output/keyword_density_fulltext_bars.png",
            dpi=150, bbox_inches="tight")
plt.show()
print("Saved → ../output/keyword_density_fulltext_bars.png")
```

## 7. Key Findings

**T1 (商业合同违约) is absent from all top-20 corpus terms.** Commercial/debt vocabulary (出借, 借条, 催款 etc.) is specialised and rarely appears in the broader case pool — confirming that T1's 23.9% corpus share comes from a concentrated subset of cases.

**T3 (运动队人事) dominates** 9 of the 20 terms, reflecting that sports personnel and wage disputes generate the broadest shared vocabulary (体育局, 运动队, 工资待遇, 人事厅, 运龄 etc.).

**Boilerplate impact († terms):** 民事, 民事判决, 民事裁定 rank #3, #4, #14 by corpus frequency despite carrying no substantive topic content — validating their addition to the modified STOPWORDS. Their topic weights are distributed across T2-T4 purely as parsing noise.

**Bridge structure:** 择业, 聘用, 体育中心, 停发, 补偿费, 微信 each span two or three topics, marking conceptual transitions: 择业 links T2↔T3↔T4 (retirement affects multiple case types); 微信 links T1↔T2 (evidence medium used in both commercial and infringement cases).
