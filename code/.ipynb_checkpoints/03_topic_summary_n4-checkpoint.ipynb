# 03 — Topic Summary (4-Topic Model)

Topic identification from the 4-topic LDA model (`output/lda_vis_chn_4.html`).

Topics are identified using pyLDAvis display ordering with loglift-ranked terms to cut through
boilerplate and surface substantive case-content signals. The pie chart mirrors the cluster
structure used in the relationship arc.

## 1. Imports & Chinese Font Setup

```python
import json, re
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.font_manager as fm

_cjk = ['PingFang SC', 'Heiti SC', 'STHeiti', 'SimHei', 'Noto Sans CJK SC']
_avail = {f.name for f in fm.fontManager.ttflist}
_font = next((f for f in _cjk if f in _avail), None)
if _font:
    matplotlib.rcParams['font.family'] = _font
    print(f'Using CJK font: {_font}')
else:
    print('No CJK font found — Chinese labels may render as boxes')
matplotlib.rcParams['axes.unicode_minus'] = False

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

## 2. Parse `lda_vis_chn_4.html`

```python
with open("../output/lda_vis_chn_4.html") as f:
    html = f.read()

match = re.search(r'var ldavis_\w+_data = ({.*?});', html, re.DOTALL)
lda_data = json.loads(match.group(1))
tinfo    = lda_data["tinfo"]
mds      = lda_data["mdsDat"]

print("topic.order (display -> gensim 1-indexed):", lda_data["topic.order"])
print("\nDisplay topic corpus shares:")
for t, f in zip(mds["topics"], mds["Freq"]):
    print(f"  Display Topic {t}: {f:.2f}%")
```

```python
# Extract top terms per display topic by loglift and frequency
topic_terms = {}
for tn in range(1, 5):
    cat = f"Topic{tn}"
    rows = [(tinfo["Term"][i], round(tinfo["Freq"][i], 2), round(tinfo["loglift"][i], 3))
            for i, c in enumerate(tinfo["Category"]) if c == cat]
    topic_terms[tn] = {
        "loglift": [t for t, f, l in sorted(rows, key=lambda x: -x[2])[:12]],
        "freq":    [t for t, f, l in sorted(rows, key=lambda x: -x[1])[:10]],
    }

for tn in range(1, 5):
    print(f"Display Topic {tn}")
    print(f"  loglift: {topic_terms[tn]['loglift']}")
    print(f"  freq:    {topic_terms[tn]['freq']}")
    print()
```

## 3. Topic Identification

The 4-topic model (`lda_vis_chn_4.html`) consolidates the five topics from the original model.
Substantive identification uses **loglift** (term exclusivity to a topic) over raw frequency,
which is inflated by procedural boilerplate (`民事`, `民事裁定`, `民事判决`).

| topic.order | Gensim idx (1-based) | Top loglift terms | Interpretation |
|---|---|---|---|
| Display T1 (31.9%) | 2 | 人事处, 离队, 招录, 体校, 违纪 | Personnel/HR & wage disputes |
| Display T2 (28.4%) | 4 | 公章, 出借, 补救措施, 专用章, 授权书 | Commercial contracts & debt |
| Display T3 (27.1%) | 3 | 共融, 停缴, 推送, 版面, 肖像权 | Retirement benefits & digital rights |
| Display T4 (12.6%) | 1 | 护理费, 椎间盘, 住宿费, 误工费, 营养费 | Medical injury compensation |

The 4-topic model merges the original T2 (image rights) and T4 (retirement/admin) into a single
Display T3, reflecting their shared theme of contested athlete entitlements.

## 4. Summary Table

```python
rows = []
for tn, m in TOPIC_META.items():
    cluster_label = (
        "Sports — Personnel/HR"    if m["sub_cluster"] == "personnel" else
        "Sports — Rights/Benefits" if m["sub_cluster"] == "rights"    else
        "Sports — Medical/Injury"  if m["sub_cluster"] == "medical"   else
        "Commercial/Financial"
    )
    rows.append({
        "Display T": f"T{tn}",
        "Corpus %": f"{m['pct']:.1f}%",
        "Cluster": cluster_label,
        "Chinese Title": m["zh"],
        "Top Terms (loglift)": ", ".join(m["top_terms_ll"][:5]),
        "English Theme": m["en"],
    })

summary = pd.DataFrame(rows).set_index("Display T")
pd.set_option("display.max_colwidth", 55)
print(summary.to_string())
summary
```

## 5. Cluster Breakdown

```python
sports_pct = sum(m["pct"] for m in TOPIC_META.values() if m["cluster"] == "sports")
comm_pct   = sum(m["pct"] for m in TOPIC_META.values() if m["cluster"] == "commercial")

sub_sums = {}
for m in TOPIC_META.values():
    sub_sums[m["sub_cluster"]] = sub_sums.get(m["sub_cluster"], 0) + m["pct"]

print("Cluster shares (4-topic model):")
print(f"  Sports-Specific Disputes  : {sports_pct:.1f}%  (T1 + T3 + T4)")
print(f"    ↳ Personnel / HR         : {sub_sums['personnel']:.1f}%  (T1)")
print(f"    ↳ Rights & Benefits      : {sub_sums['rights']:.1f}%  (T3)")
print(f"    ↳ Medical / Injury       : {sub_sums['medical']:.1f}%  (T4)")
print(f"  Commercial/Financial      : {comm_pct:.1f}%  (T2)")
print(f"  Total                     : {sports_pct+comm_pct:.1f}%")
```

## 6. Topic Distribution Pie Chart

Sports topics (T1 → T3 → T4) sit together on the left; the commercial cluster (T2) on the right.
A single outer arc labels the two main dispute clusters.

```python
COLORS = {
    1: "#1E8449",   # dark green  — personnel/HR
    2: "#7D3C98",   # purple      — commercial
    3: "#E67E22",   # orange      — rights/retirement
    4: "#2980B9",   # blue        — medical injury
}
SPORTS_COLOR = "#1A6A9E"
COMM_COLOR   = "#7D3C98"

# Sports cluster (T1, T3, T4) then commercial (T2), starting from top
topic_order  = [1, 3, 4, 2]
pcts         = [TOPIC_META[t]["pct"] for t in topic_order]
wedge_labels = [f"T{t}\n{TOPIC_META[t]['pct']:.1f}%" for t in topic_order]

fig, ax = plt.subplots(figsize=(10, 9))
ax.set_aspect("equal")
ax.set_facecolor("#f8f9fa")
fig.patch.set_facecolor("#f8f9fa")

wedges, texts = ax.pie(
    pcts,
    labels=wedge_labels,
    colors=[COLORS[t] for t in topic_order],
    startangle=90,
    labeldistance=1.10,
    wedgeprops=dict(edgecolor="white", linewidth=2.5, width=0.52),
    textprops=dict(fontsize=8.5, fontweight="bold"),
)
for text, tn in zip(texts, topic_order):
    text.set_color(COLORS[tn])

ax.text(0, 0, "Athlete\nCourt Cases\n(n = 79)",
        ha="center", va="center", fontsize=10.5,
        fontweight="bold", color="#2C3E50")

# ── outer category arc ──────────────────────────────────────────────────────
# Sports cluster = wedges 0–2 (T1, T3, T4); commercial = wedge 3 (T2)
sports_t1 = wedges[0].theta1   # start of T1 at 90°
sports_t2 = wedges[2].theta2   # end of T4
comm_t1   = wedges[3].theta1   # start of T2
comm_t2   = wedges[3].theta2   # end of T2

ARC_INNER, ARC_OUTER, GAP = 1.38, 1.50, 3.0

ax.add_patch(mpatches.Wedge(
    (0, 0), ARC_OUTER, sports_t1 + GAP, sports_t2 - GAP,
    width=ARC_OUTER - ARC_INNER, color=SPORTS_COLOR, alpha=0.90, edgecolor="none",
))
ax.add_patch(mpatches.Wedge(
    (0, 0), ARC_OUTER, comm_t1 + GAP, comm_t2 - GAP,
    width=ARC_OUTER - ARC_INNER, color=COMM_COLOR, alpha=0.90, edgecolor="none",
))

LABEL_R    = 1.70
sports_mid = np.radians((sports_t1 + sports_t2) / 2)
comm_mid   = np.radians((comm_t1   + comm_t2)   / 2)

ax.text(LABEL_R * np.cos(sports_mid), LABEL_R * np.sin(sports_mid),
        "Sports-Specific\nDisputes",
        ha="center", va="center", fontsize=9.5, fontweight="bold", color=SPORTS_COLOR)
ax.text(LABEL_R * np.cos(comm_mid), LABEL_R * np.sin(comm_mid),
        "Commercial /\nFinancial Disputes",
        ha="center", va="center", fontsize=9.5, fontweight="bold", color=COMM_COLOR)

ax.set_xlim(-2.1, 2.1)
ax.set_ylim(-2.2, 2.1)
ax.axis("off")

legend_handles = [
    mpatches.Patch(color=COLORS[1], label="T1 – Sports-Team Personnel Status & Wage Benefits  (31.9%)"),
    mpatches.Patch(color=COLORS[3], label="T3 – Athlete Retirement Benefits & Online Rights Disputes  (27.1%)"),
    mpatches.Patch(color=COLORS[4], label="T4 – Sports Injury & Medical Compensation  (12.6%)"),
    mpatches.Patch(color=COLORS[2], label="T2 – Commercial Contract Breach & Debt Recovery  (28.4%)"),
]
ax.legend(handles=legend_handles, loc="lower center", ncol=1,
          fontsize=8, framealpha=0.85, edgecolor="#ccc",
          bbox_to_anchor=(0.5, -0.01))

ax.set_title(
    "LDA Topic Distribution — Athlete Court Cases (4-Topic Model)\n"
    "(outer arc = dispute cluster; slice area ∝ corpus share)",
    fontsize=11, fontweight="bold", pad=14,
)

plt.tight_layout()
plt.savefig("../output/topic_pie_n4.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved -> ../output/topic_pie_n4.png")
```
